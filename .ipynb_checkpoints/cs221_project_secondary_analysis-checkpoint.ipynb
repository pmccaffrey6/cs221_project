# Could we Help with Sparsity through Embedding?

```python
input_df
```

```python
proteins = input_df[['uniprot_ids','aa_sequence']].drop_duplicates()
```

```python
def breakapart(word): 
    return [char for char in word]
```

```python
sequences = [breakapart(i) for i in proteins['aa_sequence'].values]

proteins['sequence'] = sequences
proteins['id'] = proteins['uniprot_ids']
```

```python
sgt = SGT(kappa = 10, lengthsensitive = False)
```

```python
embedding = sgt.fit_transform(corpus=proteins)
```
